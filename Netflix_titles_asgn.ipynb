```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/netflix_titles.csv')
```

```python
print(df.describe())
```

```python
plt.hist(df['release_year'].dropna(), bins=30, edgecolor='black', color='skyblue')
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(40, 10))
df['release_year'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 20))
plt.boxplot(df['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Year Distribution')
plt.xlabel('Frequency')
plt.ylabel('Year')
plt.show()
```

```python
sns.pairplot(df[['release_year']].dropna())
plt.show()
```

```python
correlation_matrix = df[['release_year']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='release_year', data=df, palette='Set2')
plt.title('Released year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='release_year', y='release_year', data=df, palette='muted')
plt.title('Graph')
plt.show()
```
