```python
import pandas as pd
```

IMPORT THE DATASET

```python
df = pd.read_csv('/content/sample_data/patients.csv')
```

```python
print(df)
```

CHECK ALL THE NA/NULL VALUES

```python
df.isna()
```

COUNT THE NO OF NA/NULL VALUES IN PARTICULAR ATTRIBUTES

```python
df.isnull().sum()
```

MEAN/MEDIAN/MODE OF ATTRIBUTES

```python
Glucose_Level_Mean = df["Glucose_Level"].mean()
print(Glucose_Level_Mean)
```

```python
Blood_Pressure_Median = df["Blood_Pressure"].median()
print(Blood_Pressure_Median)
```

```python
Insulin_Mean = df["Insulin"].mean()
print(Insulin_Mean)
```

```python
BMI_Mean = df["BMI"].mean()
print(BMI_Mean)
```

REPLACING THE NA/NULL VALUES WITH MEAN/MEDIAN/MODE OF ATTRIBUTES

```python
df["Glucose_Level"].fillna(Glucose_Level_Mean, inplace=True)
df["Blood_Pressure"].fillna (Blood_Pressure_Median, inplace=True)
df["Insulin"].fillna (Insulin_Mean, inplace=True)
df["BMI"].fillna (BMI_Mean, inplace=True)
```

CHECK FOR ANY MORE NA/NULL VALUES

```python
df.isnull().sum()
```

```python
df.isna()
```

PRINT DF

ALL THE MISSING DATA HAS BEEN REPLACED

```python
print(df)
```

NOW TO SCALE USING MIN MAX SCALING USING SKLEARN LIBRARY

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Glucose_Level']] = scaler.fit_transform(df[['Age', 'Glucose_Level']])
```

```python
df['Blood_Pressure'] = (df['Blood_Pressure'] - df['Blood_Pressure'].min()) / (df['Blood_Pressure'].max() - df['Blood_Pressure'].min())
```

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['BMI']] = s_scaler.fit_transform(df[[ 'BMI']])
```

```python
df['BMI']
```

```python
df['Insulin'] = (df['Insulin'] - Insulin_Mean) / df['Insulin'].std()
```
